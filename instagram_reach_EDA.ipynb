## Since we are working on a dataset based on Instagram Reach
we should start by exploring the feature that contains data about reach. In our data, the Impressions column contains the data about the reach of an Instagram post. So let’s have a look at the distribution of the Impressions:

```python
fig = px.histogram(df,
                   x="Impressions",
                   nbins=10,
                   title="Distribution of impressions")
fig.show()
```

## The distribution of reach from different sources:

```python
reach_sources = ['From Home', 'From Hashtags', 'From Explore', 'From Other']
reach_counts = [df[source].sum() for source in reach_sources]

colors = ['#E74C3C', '#76D7C4 ', '#F4D03F', '#AED6F1']

fig = px.pie(data_frame=df, names=reach_sources, values=reach_counts,
             title = "Reach From Different Sources", color_discrete_sequence=colors)

fig.show()
```

## let’s have a look at all the metrics like Likes, Saves, and Follows from each post over time:

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=df.index, y=df['Likes'], name='Likes'))
fig.add_trace(go.Scatter(x=df.index, y=df['Saves'], name='Saves'))
fig.add_trace(go.Scatter(x=df.index, y=df['Follows'], name='Follows'))

fig.update_layout(title='Metrics Over Time',
                  xaxis_title='Date',
                  yaxis_title='Count')

fig.show()
```

## Explore the hashtags column in detail.
Each post contains different combinations of hashtags, which impacts reach on Instagram. So let’s have a look at the distribution of hashtags to see which hashtag is used the most in all the posts:

```python
# Create a list to store all hashtags in our data frame
all_hashtags = []

# Iterate through each row in the 'Hashtag' column
for row in df['Hashtags']:
  hashtags = str(row).split()
  hashtags = [tag.strip() for tag in hashtags]
  all_hashtags.extend(hashtags)

# Create a pandas series to store the hashtags distribution
hashtags_distribution = pd.Series(all_hashtags).value_counts().reset_index()
hashtags_distribution.columns = ['Hashtag', 'Count']

fig = px.bar(hashtags_distribution, x="Hashtag", y="Count",
             title = "Distribution of Hashtags")

fig.show()

```

## Understand the distribution of likes and impressions received from the presence of each hashtag on the post:

```python
# Create a dictionary to store the likes and impressions for each hashtag
hashtag_likes = {}
hashtag_impressions = {}

# Iterate over each row in the dataset
for index, row in df.iterrows():
    hashtags = str(row['Hashtags']).split()
    for hashtag in hashtags:
        hashtag = hashtag.strip()
        if hashtag not in hashtag_likes:
            hashtag_likes[hashtag] = 0
            hashtag_impressions[hashtag] = 0
        hashtag_likes[hashtag] += row['Likes']
        hashtag_impressions[hashtag] += row['Impressions']

# Create a DataFrame for likes distribution
likes_distribution = pd.DataFrame(list(hashtag_likes.items()),
                                  columns=['Hashtag', 'Likes'])

# Create a DataFrame for impressions distribution
impressions_distribution = pd.DataFrame(list(hashtag_impressions.items()),
                                        columns=['Hashtag', 'Impressions'])

fig_likes = px.bar(likes_distribution, x='Hashtag', y='Likes',
                   title='Likes Distribution for Each Hashtag')

fig_impressions = px.bar(impressions_distribution, x='Hashtag',
                         y='Impressions',
                         title='Impressions Distribution for Each Hashtag')

fig_likes.show()
fig_impressions.show()
```
